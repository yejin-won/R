```r
ta <- read.table("./Data/trafficAccident_2011_2019.txt", header = T, stringsAsFactors = F)
head(ta)
```

```r
str(ta)
```

```r
colnames(ta)[5:16] <- substring(colnames(ta)[5:16], 2, length(colnames(ta)[5:16]))
head(ta)
```

```r
ta <- subset(ta,!자치구별=='합계')
```

```r
rownames(ta) <- c()
```

```r
head(ta)
tail(ta)
```

```r
ta <- ta[,-4]
head(ta)
```

```r
library(stringr)
```

```r
for(i in 4:15) {
    
    ta[,i] <- str_replace_all(ta[,i], "-", "0")
    
}

head(ta)
```

```r
ta1 <- ta
```

```r
head(ta1)
```

```r
발생건수 = c()
사망자수 = c()
부상자수 = c()
for(k in c('발생건수','사망자수','부상자수')){
    sub <- subset(ta1,구분==k)
    for(j in 1:nrow(sub)){
        for(i in 4:15){
            if(k=='발생건수'){
                발생건수 <- rbind(발생건수,sub[j,i])
            }else if(k=='사망자수'){
                사망자수 <- rbind(사망자수,sub[j,i])
            }else{
                부상자수 <- rbind(부상자수,sub[j,i])
            }
        }
    }
}
```

```r
seoul_new <- data.frame(년도=rep(2011:2019,each=12),월=rep(1:12,9),
                        자치구명=rep(unique(ta1$자치구별),each=12),발생건수,사망자수,부상자수)
```

```r
seoul_new
```

```r
seoul_new$발생건수 <- as.integer(seoul_new$발생건수)
seoul_new$사망자수 <- as.integer(seoul_new$사망자수)
seoul_new$부상자수 <- as.integer(seoul_new$부상자수)
str(seoul_new)
```
