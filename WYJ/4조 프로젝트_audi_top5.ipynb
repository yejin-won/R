```r
test.sc <- as.data.frame(audi_top5[,-audi_top5$price])
```

```r
test.sc$price <- audi_top5$price
```

```r
test.sc
```

```r
set.seed(1234)
library(caret)
```

```r
samp <- createDataPartition(test.sc$price,p=0.7,list=F)
data.tr <- test.sc[samp,]
data.test <- test.sc[-samp,]
```

```r
x <- subset(data.test,select = -price)
y <- data.test$price
```

```r
library(nnet)
```

```r
result.nnet = c()
for(i in 1:5){
    model.nnet <- nnet(price~.,data=data.tr,size=3)
    pred.nnet <- predict(model.nnet,x,type="raw")
    result.nnet <- append(result.nnet,mean(pred.nnet==y))
}
mean.nnet <- mean(result.nnet)
paste("인공신경망 : ",mean.nnet)
```

```r
library(party)
```

```r
# 2) 의사결정나무

result.ctree = c()
for(i in 1:5){
    model.ctree <- ctree(price~.,data=data.tr)
    pred.ctree <- predict(model.ctree,x)
    result.ctree <- append(result.ctree,mean(pred.ctree==y))
}
mean.ctree <- mean(result.ctree)
paste("의사결정나무 : ",mean.ctree)
```

```r
library(randomForest)
```

```r
result.rf = c()
for(i in 1:5){
    model.rf <- randomForest(price~.,data=data.tr,ntree=100)
    pred.rf <- predict(model.rf,x)
    result.rf <- append(result.rf,mean(pred.rf==y))
}
mean.rf <- mean(result.rf)
paste("랜덤포레스트 : ",mean.rf)
```

```r
library(e1071)
```

```r
result.svm = c()
for(i in 1:5){
    model.svm <- svm(price~.,data=data.tr)
    pred.svm <- predict(model.svm,x)
    result.svm <- append(result.svm,mean(pred.svm==y))
}
mean.svm <- mean(result.svm)
paste("SVM : ",mean.svm)
```

```r
result.logR = c()
for(i in 1:5){
    model.logR <- multinom(price~.,data=data.tr)
    pred.logR <- predict(model.logR,x)
    result.logR <- append(result.logR,mean(pred.logR==y))
}
mean.logR <- mean(result.logR)
paste("다중로지스틱회귀분석 : ",mean.logR)
```
