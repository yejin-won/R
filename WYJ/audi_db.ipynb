```r
A3 <- read.csv("./Data/A3_opt.csv")
Q3 <- read.csv("./Data/Q3_opt.csv")
A4 <- read.csv("./Data/A4_opt.csv")
```

```r
head(A3)
head(Q3)
head(A4)
```

```r
library(dplyr)
```

```r
# transmission 컬럼 생성 
A3$transmission <- ifelse(A3$Automatic == 1, 'Automatic', ifelse(A3$Manual == 1, 'Manual','Semi_Auto'))
Q3$transmission <- ifelse(Q3$Automatic == 1, 'Automatic', ifelse(Q3$Manual == 1, 'Manual','Semi_Auto'))
A4$transmission <- ifelse(A4$Automatic == 1, 'Automatic', ifelse(A4$Manual == 1, 'Manual','Semi_Auto'))
```

```r
# fuelType 컬럼 생성 
A3$fuelType <- ifelse(A3$fuelType == 1, 'Petrol','Diesel')
Q3$fuelType <- ifelse(Q3$fuelType == 1, 'Petrol','Diesel')
A4$fuelType <- ifelse(A4$fuelType == 1, 'Petrol','Diesel')
```

```r
#필요한 컬럼 남기고 삭제 
A3 <- A3[,c(1,2,3,4,5,6,10)]
Q3 <- Q3[,c(1,2,3,4,5,6,10)]
A4 <- A4[,c(1,2,3,4,5,6,10)]
```

```r
# 연식별 평균 가격 
A3연식 <- aggregate(price ~ age, A3, mean)
Q3연식 <- aggregate(price ~ age, Q3, mean)
A4연식 <- aggregate(price ~ age, A4, mean)
```

```r
# 연료타입별 평균가격 
A3연료 <- aggregate(price ~ fuelType, A3, mean)
Q3연료 <- aggregate(price ~ fuelType, Q3, mean)
A4연료 <- aggregate(price ~ fuelType, A4, mean)
```

```r
# 변속기별 평균가격 
A3변속기 <- aggregate(price ~ transmission, A3, mean)
Q3변속기 <- aggregate(price ~ transmission, Q3, mean)
A4변속기 <- aggregate(price ~ transmission, A4, mean)
```

```r
# 컬럼 이름 바꾸기 
colnames(A3연식) <- c('value','price')
colnames(A3연료) <- c('value','price')
colnames(A3변속기) <- c('value','price')
colnames(Q3연식) <- c('value','price')
colnames(Q3연료) <- c('value','price')
colnames(Q3변속기) <- c('value','price')
colnames(A4연식) <- c('value','price')
colnames(A4연료) <- c('value','price')
colnames(A4변속기) <- c('value','price')
```

```r
# 3개 df 결합 
A3_db <- rbind(rbind(A3연식, A3연료), A3변속기)
Q3_db <- rbind(rbind(Q3연식, Q3연료), Q3변속기)
A4_db <- rbind(rbind(A4연식, A4연료), A4변속기)
```

```r
A3_db
```

```r
# model 컬럼 생성 
A3model <- rep("A3", nrow(A3_db))
Q3model <- rep("Q3", nrow(Q3_db))
A4model <- rep("A4", nrow(A4_db))
model <- c(A3model, Q3model, A4model)
```

```r
# feature 컬럼 생성 
A3feature <- c(rep('age', (nrow(A3_db)-5)),rep('fuelType', 2) , rep('Transmission', 3))
Q3feature <- c(rep('age', (nrow(Q3_db)-5)),rep('fuelType', 2) , rep('Transmission', 3))
A4feature <- c(rep('age', (nrow(A4_db)-5)),rep('fuelType', 2) , rep('Transmission', 3))
feature <- c(A3feature, Q3feature, A4feature)
```

```r
# 최종 df 만들기 
audi_db <- cbind(model, feature, rbind(rbind(A3_db, Q3_db), A4_db))
```

```r
audi_db
```

```r
# df 추출 
write.csv(audi_db, "./Data/audi_db.csv", row.names = F)
```
