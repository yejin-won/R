# 회귀분석(Regression)
- 데이터를 이용하여 함수식(Y=aX+b)를 구한다.
- 기울기(DL :Weight)와 절편(DL:Bias)을 DataSet으로 최적값을 구한다. (정답과 오차율의 차이가 제일 적은 것으로 선택--> 인공신경망으로 발전)
- 기울기와 절편으로 Y값을 예측할 수 있다.

### 단순회귀 분석 : Y = aX + b

```r
# 기본 통계량 확인

summary(regression) # age가 적은쪽의 데이터들이 많으니 왼쪽으로 쏠린 그래프가 나옴
```

```r
plot(regression$Height,regression$Weight)
```

```r
plot(regression$Weight~regression$Height)

# 회귀선(중심선,평균선)
r <- lm(regression$Weight~regression$Height)
abline(r,col="red")
```

```r
# 분석결과

summary(r)
```

p-value: < 2.2e-16 < 0.05     
Adjusted R-squared:  0.9333 (예측력) --> 93% 정확도가 있다.    
Residual standard error: 6.355 (표준오차)   

```r
# *** 중요하다는 표시
# y = a * x + b
# intercept ---> b

# 회귀식
r
```

```r
# 키가 170인 사람의 몸무게
# a : 0.6147 , b : -40.8659 

0.6147 * 170 + (-40.8659)

# 표준오차를 적용한 키가 170인 사람의 몸무게

cat(0.6147 * 170 + (-40.8659)-6.355,"~",0.6147 * 170 + (-40.8659)+6.355)
```

# IQ

```r
iq <- read.csv("./Data/IQ.csv")
head(iq)
tail(iq)
```

```r
# IQ가 126인 사람의 성적을 예측해 보시오
```

```r
options(warn =-1)
par(family = "AppleGothic")
plot(iq$IQ,iq$성적)

l <- lm(iq$성적~iq$IQ)
abline(l,col="red")
```

```r
l

summary(l)
```

```r
# IQ가 126인 사람의 성적
0.6714 *126 

# 표준오차를 적용한 IQ가 126인 사람의 성적
cat(0.6714 *126 -3.322,"~",0.6714 *126 +3.322)

# 위의 키-몸무게 데이터와 달리 summary(l)의 a,b의 중요도 중 b는 표시가 안되어있으므로 의미없는 데이터를 의미
# 그래서 y=ax로 예측해야함 만약 반대로 a의 중요도가 표시가 안되어있다면 그건 데이터 자체가 잘못되었다는 것을 의미
```

```r
# 성적과 가장 관련이 깊은 항목은?

cor(iq[,-1])
```

```r
# 학원을 몇군데 다녀야 성적이 90점 이상이 될까?

options(warn =-1)
par(family = "AppleGothic")
plot(iq$성적,iq$다니는.학원수)

k <- lm(iq$다니는.학원수~iq$성적)
abline(k,col="red")
```

```r
k

summary(k)
```

```r
# 성적이 90점 이상이 되려면 다녀야 하는 학원수
0.12410*90 + (-7.89163)

# 표준편차 적용된 성적이 90점 이상이 되려면 다녀야 하는 학원수
cat(0.12410*90 + (-7.89163)-0.7882,"~",0.12410*90 + (-7.89163)+0.7882)
```

```r
# 성적이 85점이라면 IQ는 얼마?

options(warn =-1)
par(family = "AppleGothic")
plot(iq$성적,iq$IQ)

s <- lm(iq$IQ~iq$성적)
abline(s,col="red")
```

```r
s

summary(s)
```

## 다중 회귀 분석
- 여러개의 항목들로 구성된 식으로 예측하는 방법

### attitude
- 대형 금융기관의 사무직 대상으로 진급의 요소에 대한 설문조사 30개부서(무작위) 약 35명을 대상으로 한 설문지 데이터
- 각 수치는 각 부서의 7개의 질문에 대한 점수를 나타낸다

```r
help(attitude)
```

```r
head(attitude)
```

```r
str(attitude)
```

```r
plot(attitude)
```

```r
# 회귀 분석
model <- lm(rating~.,data=attitude) # . : 앞에 있는 target column을 제외한 나머지 출력 
model
```

rating =  0.61319 * complaints + (-0.07305) * privileges+..... + 10.78708 

```r
# Model 검토
summary(model)
```

> Adjusted R-squared:  0.6628 

```r
cor(attitude)
```

```r
# 예측력을 높이기 위해 상대적으로 기여도가 낮은 변수를 제거

reduced <- step(model,direction = "backward")
```

```r
summary(reduced)
```

> Adjusted R-squared:  0.6864 (6개있을 때보다 예측력이 증가함)
