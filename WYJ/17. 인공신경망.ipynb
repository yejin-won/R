```r
print(head(iris))
```

```r
# 학습용 데이터와 검증용 데이터의 분리

set.seed(1234) #샘플링할때 번호 동일하게 만들기 위한것
samp <- c(sample(1:50,35), sample(51:100,35), sample(101:150,35))
samp
```

```r
# 학습용 데이터셋
data.tr <- iris[samp,]
```

```r
# 검증용 데이터셋
data.test <- iris[-samp,]
```

```r
summary(data.tr)
summary(data.test)
```

### 학습시키기

```r
# 인공 신경망 모델 사용
library(nnet)
```

```r
# Model 만들기
model.nnet <- nnet(Species ~ .,data=data.tr, size=3) 
# size : 은닉층 갯수로 sample 수보다 작아야 함 (Hyper Parameter)
```

```r
# 문제 및 정답 만들기
x <- subset(data.test,select=-Species)
y <- data.test$Species
```

```r
# 검증

pred <- predict(model.nnet, x, type = "class") # class : 분류된것(객관식), raw : 주관식
head(pred)
```

```r
# 정확도

mean(pred == y) # 95% 정확함
```

```r
# 정오표 확인

table(pred,y)
```

# 학습시킨 머신러닝 사용하기

```r
# Sepal.Length = 5.3
# Sepal.Width = 3.3
# Petal.Length = 1.6
# Petal.Width = 0.3
# 인 꽃의 품종은?
```

- 현재 AI는 105개의 데이터로 공부한 모델.
- Training과 Test를 통해서 정확도가 입증되었다면 전체 Dataset으로 AI로 구성하여 사용

```r
# 사용가능하다고 판단되면 모든 데이터를 기준으로 학습
model.nnet <- nnet(Species ~ ., data=iris, size = 3)
```

```r
flower <- data.frame(Sepal.Length = 5.3,Sepal.Width = 3.3,Petal.Length = 1.6,Petal.Width = 0.3)
#data.frame 말고 list도 사용 가능
predict(model.nnet, flower, type = "class")
```

```r
# Sepal.Length = 9.3
# Sepal.Width = 2.2
# Petal.Length = 7.5
# Petal.Width = 3.5
# 인 꽃의 품종은?
```

```r
flower2 <- data.frame(Sepal.Length = 9.3,Sepal.Width = 2.2,Petal.Length = 7.5,Petal.Width = 3.5)
# data.frame 말고 list도 사용 가능
# iris의 범위를 넘으면 제일 큰 virginica로 나옴!!!
# 이거슨 iris의 데이터 범위 이상이므로 iris가 아님~!!!
predict(model.nnet, flower2, type = "class")
```
