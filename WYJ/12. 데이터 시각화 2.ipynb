### 한 화면에 차트 여러개 그려서 비교하기

```r
price <- runif(10,1,100)
plot(price)
```

```r
# 2행 2열 차트 그리기 : type 속성으로 구분해보기 

par(mfrow = c(2,2))
plot(price, type ="l")
plot(price, type ="o")
plot(price, type ="h")
plot(price, type ="s")
```

```r
# 2행 2열 차트 그리기 : pch 속성으로 구분해보기 

par(mfrow = c(2,2))
plot(price,type="o",pch=5)
plot(price,type="o",pch=15)
plot(price,type="o",pch=20,col="blue")
plot(price,type="o",pch=c(rep(15,8),20,20),col="orange",cex = c(rep(1,8),3,1))
```

### 중첩 자료 시각화

```r
x <- c(1,2,3,4,2,4)
y <- rep(2,6)
```

```r
plot(x,y)
```

```r
# Method 1) 문제점 => 데이터 가림

xy.df <- as.data.frame(table(x,y))
xy.df
plot(x,y,
     pch =15,
     cex =2 * xy.df$Freq
    )
```

```r
options(warn =-1)
par(family = "AppleGothic")

# 기타 그래프
x <- c(1,3,6,8,9)
y <- c(12,56,78,32,9)

plot(x,y)

# 선분(점과 점을 잇는 선) 그리기 (3,3)과 (4,4)를 연결
# (1,1) = (1,12), ....(4,4)=(8,32)...
segments(6,78,8,32,col="red")

# 화살표 그리기
arrows(3,56,1,12,col="blue")

# 사각형 그리기
rect(4,20,6,30,col="red",border="red",density = 10)

# 글자 쓰기
text(4,40,"This is a 샘플",srt=50)

# 테두리 활용하기
mtext("character",side=4,adj=0.5)

# chart의 테두리 그리기
box(lty=1,col="red")

# x축의 추가 :y축의 40 위치에 0~10까지 표시하고, 색은 빨간색
axis(1,pos=40,at=0:10,col="red")

# y축의 추가 : x축의 6위치에 20-~60까지 표시하고 색은 파란색
axis(2,pos=6,at=20:60,col="blue")
```

```r
Elec1 <-arrange(Elec,desc(pop))
Elec1$합계 <- apply(Elec1[,5:7],1,sum)
Elec1 <- as.matrix(head(Elec1,5)[,4:8])
Elec1 <- t(Elec1)
Elec1
```

```r
Elec_N <- as.matrix(head(arrange(Elec,desc(pop)),5)[,2:3])
Elec_N
colnames(Elec1) <- paste(Elec_N[,1],Elec_N[,2])
rownames(Elec1) <- c("투표수","문재인","홍준표","안철수","합계")

Elec1
```

```r
options(scipen=100)
par(family = "AppleGothic")

barplot(Elec1[2:4,],
        ylim=c(0,500000),
        col = c("lightblue","red","lightgreen"),
        cex.name = 0.7,
        legend.text = T
       )
for(i in 1:5){
    text(0.7+(1.2*(i-1)),Elec1[2,i]/2,paste(round(Elec1[2,i] / Elec1[5,i] * 100, 2), "%"))
    text(0.7+(1.2*(i-1)),Elec1[2,i]+Elec1[3,i]/2,
         paste(round(Elec1[3,i] / Elec1[5,i] * 100, 2), "%"))
    text(0.7+(1.2*(i-1)),Elec1[2,i]+Elec1[3,i]+Elec1[4,i]/2,
         paste(round(Elec1[4,i] / Elec1[5,i] * 100, 2), "%"))
}
```

# 수학 함수의 시각화

y = X^2

```r
f <- function(x){
    y=x^2
    return(y)
}
```

```r
x <- 1:3
f(x)
```

```r
x <- seq(-5,5,0.1)


plot(x,f(x),
     type = "l"
    )
```

y = e^(-x) -4x^2+x+3|x|-3

```r
f <- function(x){
    y = exp(-x)-4*x^2+x+3*abs(x)-3
}
```

```r
plot(x,f(x),
     type = "l"
    )
```

```r
# y = x^2 + 2x + 3

f <- function(x){
    y = x^2 + 2*x +3
}
```

```r
plot(x,f(x),
     type = "l"
    )
```
