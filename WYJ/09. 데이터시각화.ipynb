```r
# 수직선 그래프

y <- c(1,3,2,4,8,5,7,6,9)
plot(y,
    main = "Simple Time Series",
    xlim = c(0,length(y)),
    ylim = c(0,length(y)),
    type = "h",
    col = c(rep("red",3),"blue",rep("green",6))
    )
```

```r
# 왼쪽 값을 기초로한 계단 모양 그래프

y <- c(1,3,2,4,8,5,7,6,9)
plot(y,
    main = "Simple Time Series",
    xlim = c(0,length(y)),
    ylim = c(0,length(y)),
    type = "s"
    )
```

```r
# 오른쪽 값을 기초로한 계단 모양 그래프

y <- c(1,3,2,4,8,5,7,6,9)
plot(y,
    main = "Simple Time Series",
    xlim = c(0,length(y)),
    ylim = c(0,length(y)),
    type = "S"
    )
```

```r
# 그래프 라인 타입 변경하기

y <- c(1,3,2,4,8,5,7,6,9)
plot(y,
    main = "Simple Time Series",
    xlim = c(0,length(y)),
    ylim = c(0,length(y)),
    type = "S",
    lty = 1 # 0~6 (7부터는 6으로 나눈 나머지 값이랑 동일한 선 타입이 나타남)
    )
```

```r
# 그래프 포인터 타입 변경하기 (라인그래프는 불가)

y <- c(1,3,2,4,8,5,7,6,9)
plot(y,
    main = "Simple Time Series",
    xlim = c(0,length(y)),
    ylim = c(0,length(y)),
    type = "p",
    pch = c(24,20,9),
    col = c('red','blue','yellow')
    )
```

```r
x <- runif(100)
y <- runif(100)

# pch : y>0.5 이면 1, 아니면 18
# col : y>0.5 이면 "red", 아니면 "blue"
plot(x,y,
    pch = ifelse(y>0.5,1,18),
    col = ifelse(y>0.5,'red','blue')
    )
```

## 선 그래프

```r
# 3종류 dataset의 선 그래프
# 포토샵으로 원리로 한 레이어당 선을 그리고 한꺼번에 보는 방식임
# 한꺼번에 보기위해서는 한 셀에서 실행을 해야함

abc <- c(260,300,250,280,310)
def <- c(180,200,210,190,170)
ghi <- c(210,250,260,210,270)
```

```r
par(family = "AppleGothic")
plot(abc,
    type = "o", # overlab(선과 포인터 겹쳐져서 그려짐)
    col = "red", 
    # y 축 범위가 이상하므로 limit 해야함
    ylim = c(0,400),
    axes = F, # x,y 좌표 Frame 제거
    ann = F # x,y의 label 제거
    )

# x축과 x축의 라벨을 설정
axis(1,at = 1:5, lab = c('서울','대전','대구','광주','원주')) # 1 : x축 

# y축과 y축의 라벨을 설정
axis(2,ylim = c(0,400))

# main title 지정
title(main = "과일 판매량",col.main = 'red') # plot 밖에 따로 주게 될 경우, 색깔을 바꿀수있음 (plot 안은 검정 고정)

# x축 title(label) 지정
title(xlab = '지역',col.lab = 'blue')

# y축 title(label) 지정
title(ylab = '판매량',col.lab = 'blue')

# 선그리기 함수로 def 그리기
lines(def,
      type = "o",
      pch = 22,
      col = 'green',
      lty = 2
      )

# 선그리기 함수로 ghi 그리기
lines(ghi,
      type = 'o',
      pch = 22,
      col = 'blue',
      lty = 2
      )

# 범례
legend(4,400, # x,y 좌표
       c("야구장","축구장","해변가"),
       col = c("red","green","blue"),
       pch = c(21,22,22),
       lty = c(1,2,2),
       cex = 0.8 # R에서는 기본 크기가 1임
      )
```

# 파이차트

```r
sales <- c(210,110,400,550,600,130)
pie(sales)
```

```r
par(family = "AppleGothic")
sales <- c(210,110,400,550,600,130)
week <- c("월","화","수","목","금","토")
ratio <- sales / sum(sales) *100
label <- paste(week,"\n",ratio,"%")
```

```r
par(family = "AppleGothic")

pie(sales,
    main = "영업실적 파이차트",
    labels = label,
    col = rainbow(length(sales))
   )
```

# 3D Pie Chart

```r
library(plotrix)
```

```r
par(family = "AppleGothic")
sales <- c(210,110,400,550,600,130)
week <- c("월","화","수","목","금","토")
ratio <- sales / sum(sales) *100
label <- paste(week,"\n",ratio,"%")

par(family = "AppleGothic")

pie3D(sales,
    main = "영업실적 파이차트",
    labels = label,
    explode = 0.05,
    labelcex = 0.8
   )
```
