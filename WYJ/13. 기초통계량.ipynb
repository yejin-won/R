### 기본 통계량으로 분석해보기(summary에서 나오는 항목들) 예)평균, 중앙값 ....

```r
# 밀가루 사용

flour <- c(3,-2,-1,0,1,-2)

# 다이어트 약 

diet <- c(-4,1,-3,-5,-2,-8)
```

```r
total <- c(flour,diet)
total
```

```r
# 기본 통계량

summary(total)
```

```r
# 히스토그램

hist(total)
```

```r
# 이상치 찾기 => 없음

boxplot(total)

boxplot(flour,diet)
```

```r
# 상관계수 => 둘은 관계가 없음(밀가루로 다이어트 효과가 난 것은 우연이라고 생각할 수 있음)
# 상관계수를 구할 때 데이터의 순서는 중요함!!!

cor(flour,diet)
```

```r
x <- c(1,9,3) # c(1,3,9)일 경우 => 상관계수 90%대 나옴 ----> 데이터의 순서 중요한 예시
y <- c(10,20,30)
```

```r
par(mfrow = c(1,2))
plot(x, type = "l")
plot(y,type = "l")
```

```r
cor(x,y)
```

```r
cafe <- read.csv("./Data/coffee.csv")
head(cafe)
```

```r
# 커피의 빈도수

table(cafe$Coffee)
```

```r
# 범위 빈도수

coffee_Range <- seq(0,60,10) # 범위
table(cut(cafe$Coffee,breaks = coffee_Range, right = F))#coffee를 range에 의해 짜른다
```

```r
# 줄기-잎 그림

stem(cafe$Coffee)
```

```r
# 커피와 과일주스의 변동계수 구하기 
Coffee_cv <- sd(cafe$Coffee)/mean(cafe$Coffee) 
Coffee_cv

Fruit_cv <- sd(cafe$Fruit.Cup) / mean(cafe$Fruit.Cup)
Fruit_cv
```
