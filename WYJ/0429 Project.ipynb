```r
seoul <- read.csv("./Data/newSeoul2019.csv")
```

```r
year <- function(n){
    subset(seoul,seoul$년도==n)
}
```

```r
y2011 <- year(2011)
y2012 <- year(2012)
y2013 <- year(2013)
y2014 <- year(2014)
y2015 <- year(2015)
y2016 <- year(2016)
y2017 <- year(2017)
y2018 <- year(2018)
y2019 <- year(2019)
```

```r
options(warn=-1)
par(family = "AppleGothic")
yearAcc.act <- aggregate(발생건수 ~ 년도,seoul,sum)
bp <- barplot(yearAcc.act$발생건수,
        main="서울시 연도별 교통사고 발생 건 수",
       xlab="연도(year)",
       ylab="총 발생건수",
       names = yearAcc.act$년도,
        ylim = c(0,50000),
        col = ifelse(as.numeric(yearAcc.act$발생건수)==max(as.numeric(yearAcc.act$발생건수)),'red',
                    ifelse(as.numeric(yearAcc.act$발생건수)==min(as.numeric(yearAcc.act$발생건수)),'blue','grey')))
       yat=seq(0,50000,by=10000)
axis(side=2,at=yat)
text(x=bp,y=yearAcc.act$발생건수,labels=yearAcc.act$발생건수,pos=3,cex.name=0.1)
legend("topright",legend=c("최대","최소"),fill=c("red","blue"),border="white",box.lty=0,cex=1)
```

```r
options(warn=-1)
par(family = "AppleGothic")
yearAcc.inj <- aggregate(부상자수 ~ 년도,seoul,sum)
bp.inj <- barplot(yearAcc.inj$부상자수,
        main="서울시 연도별 교통사고 부상자 수",
       xlab="연도(year)",
       ylab="총 부상자 수",
       names = yearAcc.inj$년도,
        ylim = c(0,65000),
        col = ifelse(as.numeric(yearAcc.inj$부상자수)==max(as.numeric(yearAcc.inj$부상자수)),'red',
                    ifelse(as.numeric(yearAcc.inj$부상자수)==min(as.numeric(yearAcc.inj$부상자수)),'blue','grey')))
       yat=seq(0,65000,by=10000)
axis(side=2,at=yat)
text(x=bp.inj,y=yearAcc.inj$부상자수,labels=yearAcc.inj$부상자수,pos=3,cex.name=0.1)
legend("topright",legend=c("최대","최소"),fill=c("red","blue"),border="white",box.lty=0,cex=1)
```
