```r
# data.tr의 문제, 정답 만들기 

data.tr.x <- data.tr[,-1]
data.tr.y <- data.tr[,1]
```

```r
head(data.test)
```

```r
# data.test의 문제, 정답 만들기 

data.test.x <- data.test[,-1]
data.test.y <- data.test[,1]
```

```r
# 모델학습 
# 라이브러리 import 
library(caret) 
library(e1071) 
library(party) 
library(nnet) 
library(randomForest) 
```

## 다중선형회귀모델

```r
# 모델 만들기 
model_lm <- lm(price ~ ., data = data.tr)
model_lm
```

```r
summary(model_lm)
```

```r
# 트레이닝 데이터 성능 
pred_tr <- predict(model_lm, newdata = data.tr.x)
```

```r
predicted <- pred_tr
correct <- data.tr.y
```

```r
# 정확도 구하기 
r = 1 - (sum((correct - predicted)^2) / sum((correct - mean(correct))^2))
r
```

```r
# 회귀식의 평균 절대 오차 (Mean Absolute Error)
library(Metrics)
mae(correct, predicted)
```

```r
# 테스트 데이터 성능 

pred_test <- predict(model_lm, newdata = data.test.x)
```

```r
predicted <- pred_test
correct <- data.test.y
```

```r
# 정확도 구하기 
r = 1 - (sum((correct - predicted)^2) / sum((correct - mean(correct))^2))
r
```

```r
# 회귀식의 평균 절대 오차 (Mean Absolute Error)
library(Metrics)
mae(correct, predicted)
```

> 0.005차이
