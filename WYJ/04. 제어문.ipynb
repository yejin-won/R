```r
x <- 10
y <-runif(1)*100

if(x>y){
    print("x>runif()")
}
print("---End---")
```

```r
x <- 10
y <-runif(1)*100

if(x>y){
    print("x>runif()")
}else{
    print("x<=runif()")
}

print("---End---")
```

```r
x <- 10
y <-runif(1)*100

if(x>y){
    print("x>runif()")
}else if(x<y){
    print("x<runif()")
}else{
    print("x=runif()")
}
```

```r
# 홀수 짝수 구분

a <- c(8,7,6)

if(a%%2==0){
    sprintf("%i는 짝수입니다.",a)
}else{
    cat(a,"는 홀수 입니다.")
}
```

```r
a <- c(8,7,6)

ifelse(a %% 2 == 0, sprintf("%i(은)는 짝수",a), sprintf("%i(은)는 홀수",a))
ifelse(a %% 2 == 0, paste(a," : 짝수"), "홀수")
# a는 벡터가 아니므로 cat()을 사용할 수 없음 => sprintf()나 벡터를 생성하는 paste()사용
```

```r
# 사용자로부터 data를 입력 받아 입력한 숫자가 홀수인지 짝수인지 판단하여 출력하기.

num1 <- as.integer(readline("숫자를 입력하세요 : "))
```

```r
# 입력하신 숫자는 12이고 짝수입니다. 출력하기

num2 <- as.integer(readline("숫자를 입력하세요 : "))

if(num2 %% 2 == 0){
   sprintf("입력하신 숫자는 %i이고, 짝수입니다.",num2)
}else{
   sprintf("입력하신 숫자는 %i이고, 홀수입니다.",num2)
}
```

```r
# 입력하신 점수는 88이고 B학점 입니다. 출력하기

score1 <- as.integer(readline("점수를 입력하세요 : "))
if((score1 %/%10 == 10)||(score1 %/%10 == 9)){
    print("A학점입니다.")
}else if((score1 %/%10 == 8)||(score1 %/%10 == 7)){
    print("B학점입니다.")
}else if((score1 %/%10 == 6)||(score1 %/%10 == 5)){
    print("C학점입니다.")
}else{
    print("F학점입니다.")
}
```
